```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

# para integrar matplotlib en jupyter:
%matplotlib inline
```

Importamos el archivo

```python
from google.colab import drive
drive.mount('/content/drive')
DIR_TEMP = '/content/drive/MyDrive/Colab Notebooks/ventas.csv'
df= pd.read_csv(DIR_TEMP)
```

##**1)** GRÁFICA 1

```python
x=df.num_mes
y=df.total_ventas
plt.plot(x,y,c='b')
plt.axis([1,12,100000,500000])#me queda saber como poner del 100.000 y no de 50.000
plt.xlabel('Número de mes')
plt.ylabel('Ventas')
plt.title('VENTAS POR MES')
```

##**2)** MISMOS DATOS QUE EN LA ANTERIOR PERO DISTINTA APARIENCIA

```python
x=df.num_mes
y=df.total_ventas
plt.plot(x,y,"--",color='red', linewidth=3, marker='o', markeredgecolor='black',markeredgewidth=3, markersize=2)
plt.axis([1,12,100000,500000])#me queda saber como poner del 100.000 y no de 50.000
plt.xlabel('Número de mes')
plt.ylabel('Ventas')
plt.title('VENTAS POR MES')
```

#**3)**Número de unidades vendidas por mes de cada producto

```python
x=df.num_mes
a=df.crema_facial
b=df.limpiador_facial
c=df.crema_dentifrica
d=df.gel_baño
e=df.champu
f=df.crema_hidratante
plt.axis([1,12,0,18000])
plt.xlabel('Número de mes')
plt.ylabel('Unidades')
plt.title('Información de ventas')
plt.plot(x,a,marker='o',label='Crema facial')
plt.plot(x,b,marker='o',label='Limpiador facial')
plt.plot(x,c,marker='o',label='Crema dentífrica')
plt.plot(x,d,marker='o',label='Gel de baño')
plt.plot(x,e,marker='o',label='Champu')
plt.plot(x,f,marker='o', label='Crema hidratante')
plt.legend(loc='upper left',frameon= False)#frame es el marco
```

##**4)** Ventas de pasta dentífrica

```python
x=df.num_mes
y=df.crema_dentifrica
plt.scatter(x, y, label='Crema dentífrica')
plt.xlabel('Número de mes')
plt.ylabel('Unidades')
plt.title('Información de ventas')
plt.grid(linestyle='--')#para poner la cuadricula en línea discontinua
plt.legend(loc='upper left')
```

#**5)**Ventas de crema facial frente a crema limpiadora

```python
x=df.num_mes
y=df.crema_facial
z=df.limpiador_facial
plt.xlabel('Número de mes')
plt.ylabel('Unidades')
plt.title('Ventas de crema facial y limpiador facial')
plt.grid(linestyle='--')
plt.bar(x,y,width=0.20, label='Crema facial')
plt.bar(x+0.2,z,width=0.20, label='Limpiador facial')
plt.legend(loc='upper left')
```

#**6)** VENTA DE GEL DE BAÑO

```python
x=df.num_mes
y=df.gel_baño
plt.xlabel('Número de mes')
plt.ylabel('Unidades')
plt.title('Ventas de gel de baño')
plt.grid(linestyle='--')
plt.bar(x,y,width=0.80, label='Crema facial')
```

##**8)** Gráfico de tarta con las ventas totales

```python
x= df[['crema_facial','limpiador_facial','crema_dentifrica', 'gel_baño', 'champu', 'crema_hidratante']].sum()
x_redondeado= [round(p, 2) for p in x]
y=['Crema facial','Limpiador facial', 'Crema dentrifica', 'Gel de baño', 'Champú', 'Crema hidratante']
plt.pie(x_redondeado, labels=y, autopct='%1.1f%%')#el autopct me indica cuantos decimales quiero
plt.legend(loc='lower right')
plt.legend(loc='center left',bbox_to_anchor=(0.75,0.25))
#no me da exactamente los mismos porcentajes, pero por poco, o he fallado haciendo el redondeo o es que él tiene otros datos como supongo del ejercicio anterior
```

##**9)** Ventas mensuales de gel de baño y limpiador facial (mostrarlo como dos subgráficos)

```python
x=df.num_mes
y=df.gel_baño
z=df.limpiador_facial
fig, ejes = plt.subplots(2,1)
plt.xlabel('Número de mes')
plt.ylabel('Unidades vendidas',)
ejes[0].plot(y,marker='o', color='black')
ejes[0].set_title('Ventas de gel de baño')
ejes[0].set_xticklabels([])# quito los valores del eje x pero mantengo las marcas
ejes[1].plot(x,z,marker='o', color='red')
ejes[1].set_title('Ventas de limpiador facial')
ejes[1].set_xlim(0.5,12.5)
ejes[1].set_xticks(range(1, 13))
plt.show() # puse esto al final sino al marcar los valores del eje x me aparece un texto antes
```

##**10)**Realizar el gráfico de área con las ventas de todos los productos

```python
y=[df.crema_facial, df.limpiador_facial, df.crema_dentifrica, df.gel_baño, df.champu, df.crema_hidratante]
z=['Crema facial','Limpiador facial','Crema dentrífica','Gel de baño','Champú','Crema hidratante']
x=df.num_mes
plt.xlabel('Número de mes')
plt.ylabel('Unidades vendidas')
plt.title('Datos de ventas de todos los productos')
ejes[1].set_xlim(1,12)
colores=['magenta','cyan','red','black','green','yellowgreen']
plt.stackplot(x,y, labels=z, colors= colores)
plt.legend(loc='upper left')
plt.show() #sigue sin dar exactamente igual, pero creo que los datos no son exactamente iguales
```
